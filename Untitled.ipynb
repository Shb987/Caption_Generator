```python

def extract_features(directory):
    model = VGG16()
    model = Model(inputs=model.inputs, outputs=model.layers[-2].output)  # 4096-dim features
    features = {}
    print("Extracting features...")
    for img_name in os.listdir(directory):
        filename = os.path.join(directory, img_name)
        image = load_img(filename, target_size=(224, 224))
        image = img_to_array(image)
        image = np.expand_dims(image, axis=0)
        image = preprocess_input(image)
        feature = model.predict(image, verbose=0)
        img_id = img_name.split('.')[0]
        features[img_id] = feature
    return features

# Extract features and save
image_dir = 'images'
features = extract_features(image_dir)
dump(features, open('features.pkl', 'wb'))
```

```python
from pickle import dump

def load_captions(caption_file):
    captions = {}
    with open(caption_file, 'r') as f:
        lines = f.readlines()
    for line in lines:
        tokens = line.strip().split('\t')
        if len(tokens) < 2:
            continue
        img_id, caption = tokens[0].split('#')[0], tokens[1].lower()
        if img_id not in captions:
            captions[img_id] = []
        captions[img_id].append('startseq ' + caption + ' endseq')
    return captions

# Load captions
caption_file = 'Flickr8k_text.txt'
captions = load_captions(caption_file)
dump(captions, open('captions.pkl', 'wb'))
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add
import nltk
nltk.download('punkt')

# Paths
BASE_DIR = 'archive/'
IMAGE_DIR = os.path.join(BASE_DIR, 'image/')
TEXT_FILE = os.path.join(BASE_DIR, 'text_file.txt')
FEATURES_DIR = 'features/'
os.makedirs(FEATURES_DIR, exist_ok=True)

# Step 1: Load captions from text_file
def load_captions(file_path):
    captions = {}
    with open(file_path, 'r') as f:
        for line in f:
            if ':' in line:
                filename, caption = line.strip().split(':', 1)
                filename = filename.strip()
                caption = caption.strip()
                if filename in captions:
                    captions[filename].append(caption)
                else:
                    captions[filename] = [caption]
    return captions

# Step 2: Load and preprocess images
def load_image(image_path):
    img = Image.open(image_path).resize((224, 224))  # VGG16 input size
    img = np.array(img)
    if img.shape[-1] != 3:  # Ensure RGB
        img = np.stack([img] * 3, axis=-1)
    img = preprocess_input(img)
    return img

# Step 3: Extract image features using VGG16
def extract_features(image_dir, captions):
    model = VGG16()
    model = Model(inputs=model.inputs, outputs=model.layers[-2].output)  # Remove the last layer
    features = {}
    for img_name in captions.keys():
        img_path = os.path.join(image_dir, img_name)
        if os.path.exists(img_path):
            img = load_image(img_path)
            img = np.expand_dims(img, axis=0)  # Add batch dimension
            feature = model.predict(img, verbose=0)
            features[img_name] = feature
            # Save feature to disk
            np.save(os.path.join(FEATURES_DIR, f'{img_name}.npy'), feature)
    return features

# Load data
captions = load_captions(TEXT_FILE)
print(f"Loaded {len(captions)} images with captions.")
features = extract_features(IMAGE_DIR, captions)
print(f"Extracted features for {len(features)} images.")
```

```python
import os
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle

# Paths
BASE_DIR = '.'
TEXT_FILE = os.path.join(BASE_DIR, 'captions.txt')

# Load captions
def load_captions(file_path):
    try:
        df = pd.read_csv(file_path, names=['image', 'caption'], skipinitialspace=True)
        captions = {}
        for _, row in df.iterrows():
            img_name = row['image'].strip()
            caption = row['caption'].strip()
            if img_name in captions:
                captions[img_name].append(caption)
            else:
                captions[img_name] = [caption]
        return captions
    except Exception as e:
        print(f"Error loading captions: {e}")
        return {}

# Preprocess captions
def preprocess_captions(captions):
    all_captions = []
    for img_name, caps in captions.items():
        for cap in caps:
            all_captions.append(f'<start> {cap} <end>')
    
    # Tokenize
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(all_captions)
    vocab_size = len(tokenizer.word_index) + 1
    max_length = max(len(cap.split()) for cap in all_captions)
    
    # Save tokenizer
    with open('tokenizer.pkl', 'wb') as f:
        pickle.dump(tokenizer, f)
    
    return tokenizer, vocab_size, max_length, all_captions

# Execute
captions = load_captions(TEXT_FILE)
tokenizer, vocab_size, max_length, all_captions = preprocess_captions(captions)
print(f"Vocabulary size: {vocab_size}")
print(f"Max caption length: {max_length}")
print(f"Sample tokenized caption: {tokenizer.texts_to_sequences([all_captions[0]])[0]}")
```
